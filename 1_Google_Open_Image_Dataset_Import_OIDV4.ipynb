# **Preprocess the Dataset**

```python
# import nessesory library
import pandas as pd
from tqdm import tqdm
import os

# read csv file containing
os.chdir("/content/drive/MyDrive/Semester_4/SDP-II/Dataset/OIDv4_ToolKit/OID/csv_folder")
classes_data = pd.read_csv('/content/drive/MyDrive/Semester_4/SDP-II/Dataset/OIDv4_ToolKit/OID/csv_folder/class-descriptions-boxable.csv', header=None)
classes_data.head()

# Define required classes
classes = ['Person','Weapon','Vehicle']

# Get the string corresponding to the class name and store it inside a list called as class_string 
class_string = []
for i in classes:
  # Extrat the first column's element for mating classes weapons
  req_classes = classes_data.loc[classes_data[1]==i]
  string = req_classes.iloc[0][0]

  print(string)

  #Append the sting corresponding to the class name into the list
  class_string.append(string)

print(class_string)
```

```python
def write_annotations(csv_path, image_folder_path):

  # Get only required columns from annotations csv file and rid of NOT requred columns
  annotation_data = pd.read_csv(csv_path, usecols=['ImageID','LabelName','XMin','XMax','YMin','YMax'])
  print('\n Annotation data -> \n')
  print(annotation_data.head())
  

  # Get only records with matching classes 
  filtered_class_data = annotation_data.loc[annotation_data['LabelName'].isin(class_string)].copy()
  print('\n Filtered data -> \n')
  print(filtered_class_data.head)

  # iterate through all the class string and assign a class number according to the order they are appearing in the list
  for i in range(len(class_string)):
    # Store the result of this operation into class number
    filtered_class_data.loc[filtered_class_data['LabelName']==class_string[i], 'classNumber'] = i
                            
  # Calculate center x and center y values
  filtered_class_data['center x'] = (filtered_class_data['XMax'] + filtered_class_data['XMin']) / 2
  filtered_class_data['center y'] = (filtered_class_data['YMax'] + filtered_class_data['YMin']) / 2

  # calculate width and height values
  filtered_class_data['width'] = (filtered_class_data['XMax'] - filtered_class_data['XMin']) 
  filtered_class_data['height'] = (filtered_class_data['YMax'] - filtered_class_data['YMin']) 

  # generate the dataframe with YOLO required values
  YOLO_values = filtered_class_data.loc[:, ['ImageID','classNumber','center x','center y','width','height']].copy()
  print('\n YOLO_values -> \n')
  print(YOLO_values.head())
  print("\n")

  # Change the curent directry t the one where we have Images

  image_path = image_folder_path
  os.chdir(image_path)

  # Go through the all images in the directory
  count = 0
  for current_dir, dirs, files in os.walk('.'):
    # Iterating through all the files
    for f in files:
      # Check if the files extension ends with '.jpg

      if f.endswith('.jpg'):
        # Extract only the title of the image so that corresponding text file can be generated with same name as per yolo guidlines
        image_title = f[:-4]
        # Get match data
        YOLO_file = YOLO_values.loc[YOLO_values['ImageID']==image_title] 

        # Create a copy of the above dataframe to avoid any mess
        df = YOLO_file.loc[:,['classNumber','center x','center y','width','height']].copy()

        save_path = image_path + '/' + image_title + '.txt'

        # generate a text file containaing required data in YOLO format
        df.to_csv(save_path, header = False, index = False, sep=' ')
        count+=1
        print(count,"-> File ", image_title, ' successfully saved at location', image_path)
        

```

```python
# preprocess the training dataset
train_csv_path = '/content/drive/MyDrive/Semester_4/SDP-II/Dataset/OIDv4_ToolKit/OID/csv_folder/train-annotations-bbox.csv'
train_image_path = '/content/drive/MyDrive/Semester_4/SDP-II/Dataset/OIDv4_ToolKit/OID/Dataset/train/Weapon_Person_Vehicle'
write_annotations(train_csv_path, train_image_path)
```

```python
# preprocess the testing dataset
test_csv_path = '/content/drive/MyDrive/Semester_4/SDP-II/Dataset/OIDv4_ToolKit/OID/csv_folder/test-annotations-bbox.csv'
test_image_path = '/content/drive/MyDrive/Semester_4/SDP-II/Dataset/OIDv4_ToolKit/OID/Dataset/test/Weapon_Person_Vehicle'
write_annotations(test_csv_path, test_image_path)
```

```python
# preprocess the validation dataset
valid_csv_path = '/content/drive/MyDrive/Semester_4/SDP-II/Dataset/OIDv4_ToolKit/OID/csv_folder/validation-annotations-bbox.csv'
valid_image_path = '/content/drive/MyDrive/Semester_4/SDP-II/Dataset/OIDv4_ToolKit/OID/Dataset/validation/Weapon_Person_Vehicle'
write_annotations(valid_csv_path, valid_image_path)
```
