# **Convert the xml.txt files in proper nameing convention**

```python
# unzip the manualaly uploded data to the drive
!unzip "/content/drive/MyDrive/Semester_4/SDP-II/Model_2/Fire_validation.zip" -d '/content/drive/MyDrive/Semester_4/SDP-II/Model_2'
```

```python
# code to convert txt file into proper naming convention
import os

# set up the path of directory
path = "/content/drive/MyDrive/Semester_4/SDP-II/Model_2/Fire_validation/"

for root, dir, file in os.walk(path):

  # find whether xml present in file name or not 
  for filename in file:
    if filename.endswith('.xml.txt'):
      old_name = filename

      # remove 'xml' string form txt file
      new_name = filename.split('.')[::2]
      new_name = '.'.join(new_name)

      # get the full path of files 
      old_file = os.path.join(path,old_name)
      new_file = os.path.join(path,new_name)

      # rename the file with proper format
      os.rename(old_file, new_file)

      # print the message
      print(f"file renamed as {old_name}->{new_name}")
```

# **Cleaning the abnormalities in dataset**

```python
# unzip the manualy upladed dataset and store it into deive
!unzip '/content/drive/MyDrive/Semester_4/SDP-II/Model_2/Fire_Train.zip' -d "/content/drive/MyDrive/Semester_4/SDP-II/Model_2"
```

```python
# this function will check whether the each image has its corresponding txt file or not 
def count_images(folder):

  # intitilize counter as 0
  txt = 0
  img = 0

  # set up the path 
  path = os.path.join('/content/drive/MyDrive/Semester_4/SDP-II/Model_2',folder)

  for root, dir, filename in os.walk(path):

    for file in filename:
      name = file.split('.')[:-1]

      # if both image and txt file is present then move forword
      if str(name)[2:-2]+'.txt' in filename and (str(name)[2:-2]+'.jpg' in filename or str(name)[2:-2]+'.png' in filename):
        pass

      else:
        # remove extra file 
        os.remove(os.path.join(path,file))


    # count the txt file
    for file in filename:
      if file.endswith('.txt'):
        txt+=1

    # count the image files
    for file in filename:
      if file.endswith('.png') :
        img+=1

    # count the image files
    for file in filename:
      if file.endswith('.jpg') :
        img+=1

  # print the message
  print(f"For {folder} ->images = {img}, txt files = {txt}")

# Check whether the equal no. of txt file are present for imges or not
count_images("Fire_Train")
count_images("Fire_validation")
```

# **Check whether the data is proper or not**

```python
import os
# define the function to count the imges and text file in dataset
def count_images(path):
  txt = 0
  img = 0

  for root, dir, filename in os.walk(path):

    if 'Label' in dir:
      dir.remove("Label")

    for file in filename:
      if file.endswith('.txt'):
        txt+=1

    for file in filename:
      if file.endswith('.jpg') or file.endswith('.png'):
        img+=1
    
  print(f"For {path} ->images = {img}, txt files = {txt}")

# Check whether the equal no. of txt file are present for imges or not
count_images("/content/drive/MyDrive/Semester_4/SDP-II/Model_2/Dataset/train")
count_images("/content/drive/MyDrive/Semester_4/SDP-II/Model_2/Dataset/test")
count_images("/content/drive/MyDrive/Semester_4/SDP-II/Model_2/Dataset/validation")
```
